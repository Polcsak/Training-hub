### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# Assuming 'conn' is your SQLite connection
query = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df = pd.read_sql_query(query, conn)

# Group by 'Age' and apply a lambda function to calculate the median
median_df = df.groupby('Age')[["WorkWeekHrs", "CodeRevHrs"]].median().reset_index()

median_df
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
plt.stackplot(median_df["Age"], median_df["WorkWeekHrs"], median_df["CodeRevHrs"], labels=['Work Hours', 'Code Review Hours'])
plt.xlabel('Age')
plt.ylabel('Total Hours')
plt.title('Work and Code Review Hours by Age')
plt.legend(loc='upper left')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# your code goes here
# Assuming 'conn' is your SQLite connection
query = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 25 AND 30
"""

df = pd.read_sql_query(query, conn)

# Group by 'Age' and apply a lambda function to calculate the median
median_df = df.groupby('Age')[["ConvertedComp"]].median()

median_df
```

```python
import matplotlib.pyplot as plt

median_df.plot(kind="line")
plt.title('Correlation between ConvertedComp and Age')
```

## Load the dataset

The dataset is available on the IBM Cloud at the below url.


```python
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
```

Load the data available at dataset_url into a dataframe.


```python
import requests

# URL adresa datové sady
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"

# Požadavek na získání dat z URL
response = requests.get(dataset_url)

# Kontrola, zda byl požadavek úspěšný
if response.status_code == 200:
    # Uložení obsahu do lokálního souboru
    with open('survey_data_no_index.csv', 'wb') as file:
        file.write(response.content)

    print("Soubor byl úspěšně uložen jako survey_data_no_index.csv")
else:
    print("Chyba při stahování dat:", response.status_code)



```
